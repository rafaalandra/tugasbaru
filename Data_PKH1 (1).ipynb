### **KLASIFIKASI KELAYAKAN PENERIMAAN PKH**

Program Keluarga Harapan (PKH) adalah program bantuan sosial yang diberikan kepada keluarga miskin untuk meningkatkan kesejahteraan mereka. PKH merupakan program pemberian uang tunai bersyarat yang disalurkan kepada Keluarga Penerima Manfaat (KPM). Tujuan PKH adalah: Mengurangi angka kemiskinan, Memutus mata rantai kemiskinan antar-generasi, Meningkatkan kualitas sumber daya manusia, Mengubah perilaku yang kurang mendukung dalam peningkatan kesejahteraan. PKH memberikan dukungan di berbagai sektor, seperti pendidikan, kesehatan, gizi, serta kebutuhan dasar lainnya. Peserta PKH diwajibkan melaksanakan persyaratan dan ketentuan yang telah ditetapkan, seperti kehadiran di fasilitas pendidikan atau fasilitas kesehatan.

**VISUALISASI DATA**

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membuat dataframe
df = pd.DataFrame(data)

# Plot data: Jumlah orang berdasarkan Kelayakan (Penerima/Bukan penerima)
plt.figure(figsize=(8, 6))
df['Kelayakan'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Jumlah Orang Berdasarkan Kelayakan')
plt.xlabel('Kelayakan')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(df[df['Kelayakan'] == 'Penerima']['Pendapatan'], alpha=0.5, label='Penerima', bins=10)
plt.hist(df[df['Kelayakan'] == 'Bukan penerima']['Pendapatan'], alpha=0.5, label='Bukan penerima', bins=10)
plt.title('Income Distribution by Eligibility Status')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Calculate total income by eligibility
total_income = df.groupby('Kelayakan')['Pendapatan'].sum()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(total_income, labels=total_income.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'salmon'])
plt.title('Total Income Distribution by Eligibility Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Menghitung jumlah penerima dan bukan penerima berdasarkan status
summary = df.groupby(['Status', 'Kelayakan']).size().unstack(fill_value=0)

# Plot
plt.figure(figsize=(10, 6))
summary.plot(kind='bar', stacked=True, color=['lightblue', 'lightcoral'], edgecolor='black')

# Menambahkan label dan judul
plt.title('Jumlah Penerima dan Bukan Penerima Berdasarkan Status')
plt.xlabel('Status')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)  # Mengatur rotasi label pada sumbu x
plt.legend(title='Kelayakan', loc='upper right')

# Tampilkan plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat plot Strip antara Umur dan Kelayakan
plt.figure(figsize=(8, 5))
sns.stripplot(x='Kelayakan', y='Umur', data=df, jitter=True)
plt.title('Strip Plot antara Umur dan Kelayakan')
plt.xlabel('Kelayakan (Penerima: 1, Bukan Penerima: 0)')
plt.ylabel('Umur')
plt.grid(True)

# Menampilkan plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat plot Line antara Umur dan Kelayakan
plt.figure(figsize=(8, 5))
sns.lineplot(x='Umur', y='Kelayakan', data=df, marker='o')
plt.title('Line Plot antara Umur dan Kelayakan')
plt.xlabel('Umur')
plt.ylabel('Kelayakan (Penerima: 1, Bukan Penerima: 0)')
plt.grid(True)

# Menampilkan plot
plt.show()
```

```python
# Menghitung jumlah penerima berdasarkan anak sekolah
kelompok = df.groupby(['Anak Sekolah', 'Kelayakan']).size().unstack()

# Membuat grafik batang
kelompok.plot(kind='bar', stacked=True)

# Menambah judul dan label sumbu
plt.title('Korelasi antara Anak Sekolah dan Kelayakan Penerimaan')
plt.xlabel('Anak Sekolah')
plt.ylabel('Jumlah Orang')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# Kategorikan lansia
df['Lansia'] = df['Usia'] >= 60

# Hitung jumlah penerima dan bukan penerima berdasarkan status lansia
kelayakan_lansia = df.groupby(['Lansia', 'Penerima']).size().unstack(fill_value=0)

# Visualisasi data dengan line chart
kelayakan_lansia.plot(kind='line', marker='o')

# Setel judul dan label
plt.title('Korelasi antara Lansia dan Kelayakan Penerima')
plt.xlabel('Lansia')
plt.ylabel('Jumlah')
plt.xticks([0, 1], ['Non-Lansia', 'Lansia'])
plt.grid(True)
plt.show()
```

```python
# Kategorikan lansia
df['Lansia'] = df['Usia'] >= 60

# Hitung jumlah penerima dan bukan penerima berdasarkan usia (lansia dan non-lansia)
kelayakan_usia = df.groupby(['Usia', 'Penerima']).size().unstack(fill_value=0)

# Plot line chart
plt.figure(figsize=(10,6))
for status in kelayakan_usia.columns:
    plt.plot(kelayakan_usia.index, kelayakan_usia[status], marker='o', label=status)

# Menambahkan detail pada grafik
plt.title('Korelasi antara Usia dan Kelayakan Penerima')
plt.xlabel('Usia')
plt.ylabel('Jumlah Orang')
plt.legend(title='Status Kelayakan')
plt.grid(True)
plt.show()
```

```python
# Memilih kolom yang relevan untuk analisis korelasi
df_corr = df[['Pendapatan', 'Status', 'Umur', 'Anak Sekolah', 'Lansia', 'Kelayakan']]

# Menghitung matriks korelasi
corr_matrix = df_corr.corr()

# Membuat heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Variabel Terhadap Kelayakan')
plt.show()
```
